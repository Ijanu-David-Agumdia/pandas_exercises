### Assign it to a variable titanic 

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_titanic = pd.read_csv('train.csv')
df_titanic
```

### Set PassengerId as the index 

```python
df_titanic.set_index('PassengerId')
```

### Create a pie chart presenting the male/female proportion

```python
males = (df_titanic['Sex'] == 'male').sum()
females = (df_titanic['Sex'] == 'female').sum()
```

```python
x=[males, females]
```

```python
plt.pie(x,
    explode=[0.1, 0],
    labels=['Male', 'Female'],
    colors=['yellow','green'],
    autopct='%1.1f%%',
    pctdistance=0.6,
    shadow=False,
    labeldistance=1.1,
    startangle=90,
    radius=1,
    counterclock=True,
    wedgeprops=None,
    textprops=None,
    center=(0, 0),
    frame=False,
    rotatelabels=False,
    normalize=True,
    data=None,
)
plt.title('Sex Propotion')
plt.tight_layout()
```

### Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
lm = sns.lmplot(x = 'Age', y = 'Fare', data = df_titanic, hue = 'Sex', fit_reg=False)

lm.set(title = 'Fare x Age')

axes = lm.axes
axes[0,0].set_ylim(-5,)
axes[0,0].set_xlim(-5,85);
```

### How many people survived?

```python
df_titanic['Survived'].sum()
```

### Create a histogram with the Fare payed

```python
df = df_titanic.Fare.sort_values(ascending = False)
df

binsVal = np.arange(0,600,10)
binsVal

plt.hist(df, bins = binsVal)

plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Payed Histrogram')

plt.show()
```
