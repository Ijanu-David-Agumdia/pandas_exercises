### Assign it to a variable called tips

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_tips = pd.read_csv('tips.csv')
df_tips
```

### Delete the Unnamed 0 column

```python
del df_tips['Unnamed: 0']
df_tips.head()
```

### Plot the total_bill column histogram

```python
total_bill = sns.distplot(df_tips.total_bill);
total_bill.set(xlabel = 'Value', ylabel = 'Frequency', title = "Total Bill");
sns.despine()
```

### Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.jointplot(x ="total_bill", y ="tip", data = df_tips);
```

###  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.pairplot(df_tips)
```

### Present the relationship between days and total_bill value

```python
sns.stripplot(x = "day", y = "total_bill", data = df_tips, jitter=True);
```

### Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.stripplot(x = "tip", y = "day", hue = "sex", data = df_tips, jitter = True);
```

###  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x = "day", y = "total_bill", hue = "time", data =df_tips);
```

### Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
df_tips
```

```python
sns.set(style = "ticks")
g = sns.FacetGrid(df_tips, col = "time")
g.map(plt.hist, 'tip');
```

### Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
g = sns.FacetGrid(df_tips, col = "sex", hue = "smoker")
g.map(plt.scatter, "total_bill", "tip")
g.add_legend();
```
